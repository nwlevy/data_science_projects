## Correlations of Numeric Variables with Solar Adoption and Each Other

### Correlations Between Features

```python
fields=solar_fields.loc[(solar_fields['Relevant Feature']==1)&(solar_fields['Data Type']=='Numeric')]['Field'].tolist()
indep_vars_mat=solar[fields]
#fig,ax=plt.subplots(figsize=(30,30))
#ax.matshow(indep_vars_mat.corr())
correlation_matrix=indep_vars_mat.corr()


```

```python
correlations=[]
for i in correlation_matrix.index:
    for j in correlation_matrix.columns:
            if i!=j:
                correlations.append((i,j,correlation_matrix[i][j]))
sorted_correlations=sorted(correlations,reverse=True,key=lambda k: k[2])
top_correlations=sorted_correlations[0:100]
print('top correlations: ')
for i in top_correlations:
    print(i)
print('top negative correlations: ')
for i in sorted_correlations[-100:]:
    print(i)
```

```python
print(solar[fields].shape)
print(solar[fields].replace(np.inf,np.nan).dropna().shape)
```

### Correlations with Solar Adoption

```python
expanded_fields=fields+['number_of_solar_system_per_household']
solar2=solar[expanded_fields].replace(' ',np.nan).astype('float64')
adoption_correlations=[]
for i in fields:
    adoption_correlations.append((i,solar2[[i,'number_of_solar_system_per_household']].corr()[i]['number_of_solar_system_per_household']))
sorted_correlations=sorted(adoption_correlations,reverse=True,key=lambda k: k[1])
print('top positive correlations: ')
print(sorted_correlations[0:20])
print('top negative correlations: ')
print(sorted_correlations[-20:])
```

```python
top_correlations=sorted_correlations[0:5]
index=np.arange(5)
correlations=[i[1] for i in top_correlations]
labels=[i[0] for i in top_correlations]

plt.figure(figsize=(15,5))
plt.bar(index,correlations)
plt.xticks(index,labels)
plt.title('Top Positive Correlations with Solar Adoption')
plt.ylabel('Correlation')

plt.show()
```

```python
bottom_correlations=sorted_correlations[-5:]
index=np.arange(5)
correlations=[i[1] for i in bottom_correlations]
labels=[i[0] for i in bottom_correlations]

plt.figure(figsize=(15,5))
plt.bar(index,correlations,color='r')
plt.xticks(index,labels)
plt.title('Top Negative Correlations with Solar Adoption')
plt.ylabel('Correlation')

plt.show()
```

```python
fig,ax=plt.subplots(2,1,figsize=(15,10))

bottom_correlations=sorted_correlations[-5:]
top_correlations=sorted_correlations[0:5]

index=np.arange(5)

top_corrs=[i[1] for i in top_correlations]
top_labels=[i[0] for i in top_correlations]

bottom_corrs=[i[1] for i in bottom_correlations]
bottom_labels=[i[0] for i in bottom_correlations]


ax[0].bar(index,top_corrs)
ax[0].set_xticks(index)
ax[0].set_xticklabels(top_labels)
ax[0].set_title('Top Positive Correlations with Solar Adoption')
ax[0].set_ylabel('Correlation')


ax[1].bar(index,bottom_corrs,color='r')
ax[1].set_xticks(index)
ax[1].set_xticklabels(bottom_labels)
ax[1].set_title('Top Negative Correlations with Solar Adoption')
ax[1].set_ylabel('Correlation')

plt.show()
```

### Observations

* electricity prices are highly correlated
* temperature metrics are highly correlated
* solar radiation is correlated with higher temperatures, but the correlation coefficient is only .75.
* elevation and atmospheric pressure are very negatively correlated
* solar adoption is positively correlated with home value, income, solar radiation, solar incentives and electricity rates.
* solar adoption is negatively correlated with relative humidity, residential electricity consumption

# Fill in Missing Latitude/Longitude

```python
census_table=pd.read_csv('../2017_Gaz_tracts_national.txt',sep='\t')
```

```python
census_table.head()
```

```python
census_table.loc[census_table['GEOID']==47143975402]
```

```python
merged_data=pd.merge(solar[['fips','lat','lon']],census_table,left_on=['fips'],right_on=['GEOID'],how='left')
merged_data.columns=list(merged_data.columns[0:10])+['INTPTLONG']
#merged_data.loc[np.isnan(merged_data['lon'])]
```

```python
def fill_in_val(x,y):
    if np.isnan(x):
        return y
    else:
        return x

merged_data['latitude']=merged_data.apply(lambda x: fill_in_val(x['lat'],x['INTPTLAT']),axis=1)    
merged_data['longitude']=merged_data.apply(lambda x: fill_in_val(x['lon'],x['INTPTLONG']),axis=1)
```
