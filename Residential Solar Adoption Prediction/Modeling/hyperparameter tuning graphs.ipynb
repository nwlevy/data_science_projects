```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hyperparameter_results=pd.read_csv('hyperparameter_tuning.csv')
hyperparameter_results.head()
```

```python
regression_frame=hyperparameter_results[['max_depth_r','n_est_r','max_depth_c','n_est_c','overall_r2_2','overall_r2_3']]
regression_frame.head()
```

```python
regression_frame['avg_r2']=pd.to_numeric((regression_frame['overall_r2_2']+regression_frame['overall_r2_3'])/2)
regression_frame.head()
```

```python
regression_frame.replace({np.nan:'No Limit'},inplace=True)
```

```python
regression_frame.dtypes
```

```python
regression_frame.groupby('max_depth_r').agg({'avg_r2':'mean'})
```

```python
regression_frame.groupby('n_est_r').agg({'avg_r2':'mean'})
```

```python
regression_frame.sort_values('avg_r2',ascending=False)
```
