```python
def bucket_local_incentives(x):
    if x<10:
        return '<10'
    elif (x>=10) & (x<20):
        return '10-19'
    elif (x>=20) & (x<30):
        return '20-29'
    elif (x>=30) & (x<40):
        return '30-39'
    elif (x>=40) & (x<50):
        return '40-49'
    elif (x>=50):
        return '50+'

    
deepsolar_original['incentive_count_bucket']=deepsolar_original['incentive_count_residential'].apply(bucket_local_incentives)
#deepsolar_original.groupby('incentive_count_bucket').agg('count')

b1=deepsolar_original.loc[deepsolar_original['incentive_count_bucket']=='<10']['owner_occupied_solar_system_density']
b2=deepsolar_original.loc[deepsolar_original['incentive_count_bucket']=='10-19']['owner_occupied_solar_system_density']
b3=deepsolar_original.loc[deepsolar_original['incentive_count_bucket']=='20-29']['owner_occupied_solar_system_density']
b4=deepsolar_original.loc[deepsolar_original['incentive_count_bucket']=='30-39']['owner_occupied_solar_system_density']
b5=deepsolar_original.loc[deepsolar_original['incentive_count_bucket']=='40-49']['owner_occupied_solar_system_density']
b6=deepsolar_original.loc[deepsolar_original['incentive_count_bucket']=='50+']['owner_occupied_solar_system_density']

all_buckets=[b1,b2,b3,b4,b5,b6]


plt.figure(figsize=(12,6))
red_square = dict(markerfacecolor='r', marker='s')
plt.boxplot(all_buckets,labels=['<10','10-19','20-29','30-39','40-49','50+'], flierprops=red_square)
plt.grid(True)
plt.title('Solar Adoption Density for Different Numbers of Residential Incentives',fontweight='bold')
plt.xlabel('Residential Incentive Count',size='large',fontweight='bold')
plt.ylabel('Solar Adoption Rate',size='large',fontweight='bold')

plt.show()
```
